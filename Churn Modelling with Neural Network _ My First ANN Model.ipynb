```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv('Churn_Modelling.csv', ) #Data set is taken from Kaggle Completition
```

```python
data.head()
```

```python
data.shape
```

```python
#We are converting categorical features into Dummy variables 
#Also storing them into a dataframe

geography=pd.get_dummies(data["Geography"],drop_first=True)
gender=pd.get_dummies(data['Gender'],drop_first=True)
```

```python
## Concatenating the Data Frames

data_new=pd.concat([data,geography,gender],axis=1)
```

```python
data_new.head()
```

```python
## Dropping Unnecessary columns from our dataset
Data_new=data_new.drop(['Geography','Gender', 'Surname', 'RowNumber', 'CustomerId'],axis=1)
```

```python
#Defining the X and Y forms 
X= Data_new.drop(['Exited'], axis = 1)
```

```python
X.head()
```

```python
y = Data_new['Exited']
```

```python
y.head()
```

```python
X.shape
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train.shape
```

```python
#Now we will be making ANN 

# Importing the Keras libraries and packages
import keras
from keras.models import Sequential #This Sequential Library is responsible for creating ANN, CNN, etc. 
from keras.layers import Dense #This Library is basically used for creating Hidden Layer
#from keras.layers import LeakyReLU,PReLU,ELU #These are activation function which I am using
from keras.layers import Dropout #Dropout is a regularization parameter in Neural Network
```

```python
# Creating  ANN
classifier = Sequential()
```

```python
# Adding the input layer and the first hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'he_uniform',activation='relu',input_dim = 11));
```

```python
# Adding the second hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'he_uniform',activation='relu'))
```

```python
# Adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'glorot_uniform', activation = 'sigmoid'))
```

```python
# Compiling the ANN
classifier.compile(optimizer = 'Adamax', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Fitting the ANN to the Training set
model_history=classifier.fit(X_train, y_train,validation_split=0.33, batch_size = 10, nb_epoch = 100)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5) 
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
# Calculate the Accuracy
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,y_test)
score
```
